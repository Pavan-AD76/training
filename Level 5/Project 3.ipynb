	
# House Price Explorer

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("house_prices.csv") 

summary_stats = df[['Price', 'SquareFootage']].describe()
print("Summary Statistics:\n", summary_stats)

df['Price_per_sqft'] = df['Price'] / df['SquareFootage']

print("\nUpdated DataFrame with Price_per_sqft:\n", df.head())

plt.figure(figsize=(8, 5))
plt.scatter(df['SquareFootage'], df['Price'], alpha=0.5, color='blue')
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.title("House Price vs. Square Footage")
plt.grid(True)

plt.show()
```

# Parallel Data Processor

```python
from dataclasses import dataclass
import pandas as pd
import matplotlib.pyplot as plt

@dataclass
class DatasetStats:
    mean: float
    max: float
    min: float
    column_name: str

    def plot_histogram(self, df):
        plt.figure(figsize=(8, 6))
        plt.hist(df[self.column_name], bins=20, alpha=0.7, color='blue', edgecolor='black')
        plt.xlabel(self.column_name)
        plt.ylabel("Frequency")
        plt.title(f"Histogram of {self.column_name}")
        plt.grid(True)
        plt.show()

file_path = "house_prices.csv" 
column_to_analyze = "Price"

df = pd.read_csv(file_path)

mean_val = df[column_to_analyze].mean()
max_val = df[column_to_analyze].max()
min_val = df[column_to_analyze].min()

data_stats = DatasetStats(mean=mean_val, max=max_val, min=min_val, column_name=column_to_analyze)

def display_menu():
    while True:
        print("\nCustom Data Class Dashboard")
        print("1. Display Stats")
        print("2. Plot Histogram")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            print(f"\n{column_to_analyze} Statistics:")
            print(f"Mean: {data_stats.mean:.2f}")
            print(f"Max: {data_stats.max:.2f}")
            print(f"Min: {data_stats.min:.2f}")
        elif choice == "2":
            data_stats.plot_histogram(df)
        elif choice == "3":
            print("Exiting dashboard.")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

display_menu()
```

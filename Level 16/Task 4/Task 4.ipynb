```
df = spark.read.option("header", "true").csv("/FileStore/tables/ipl_commentary_data.csv")
```

```
from pyspark.sql.functions import col

teams1 = df.select(col("team1_name").alias("team"))
teams2 = df.select(col("team2_name").alias("team"))

all_teams = teams1.union(teams2).distinct()

print("Number of unique teams:", all_teams.count())
display(all_teams)
```

```
df.select("match_tie_breaker").distinct().show(truncate=False)
```

```
df_completed = df.filter(col("match_status") == "completed")

df_completed.select("match_no", "match_status").show(5)
```

```
columns_to_keep = ["year", "match_no", "team1_name", "team2_name", "match_status", "match_tie_breaker"]

df_cleaned = df.select(*columns_to_keep)

display(df_cleaned)
```
